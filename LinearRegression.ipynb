# Lineer Regresyon

#Giriş - Tanım - Teori


> Lineer Regresyon, tahmine dayalı modellemede en basit ve en yayın kullanılan istatistiksel tekniklerden birdir. Gözlemlenen verilere doğrusal bir denklem uydurarak, bir bağımlı değişken y ile bir veya daha fazla bağımsız değişken  arasındaki ilişkiyi modellemeyi amaçlar.

Lineer regresyon en basit şekli, yalnızca bir bağımsı değişkenin olduğu basit lineer regresyondur. Bu durumda bir çizginin denklemi:

y = a*x+b



# Numpy ile Veri Oluşturma

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive
np.random.seed(42)
```

```python
x = 2* np.random.rand(100, 1)
y = 3 * x + 4 + np.random.rand(100, 1)
```

```python
def plot_func(a, b):
  plt.scatter(x, y, label = "Veri")
  plt.plot(x, a*x+b, label= "Denklem")
  plt.xlabel("x", fontsize = 16)
  plt.ylabel("y", fontsize = 16)
  plt.title(f"y = {a: .2f}* x +{b: .2f}")
  plt.legend()
  plt.show()
```

```python
interactive(plot_func, a=(-5, 5, 0.1), b=(-5, 5, 0.1))
```

# Numpy Verisetinde Sklearn Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(x, y)
```

a * x + b Modellenen denklem için;

*   a = model.coef_
*   b = model.intercept_

```python
model.coef_, model.intercept_
```

```python
model.predict([[1]])
```

```python
x_input = 1
model.coef_ * x_input  + model.intercept_
```

#Ev Fiyatları Verisetinin Yüklenmesi ve Linear Regression

```python
from sklearn.datasets import  fetch_openml
housing = fetch_openml(name = "house_prices", as_frame = True)
```

```python
housing.keys()
```

```python
X = housing.data
y = housing.target
```

```python
X
```

```python
X.drop(["Id"], axis = 1, inplace = True)
```

```python
X.iloc[0]
```

```python
y[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

#Veriseti Önişleme

```python
X_train.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
categories = list(X_train.select_dtypes(include = ["object"]))
```

```python
numericals = list(X_train.select_dtypes(include = ["int64", "float64"]))
```

```python
len(categories + numericals)
```

```python
from matplotlib.cbook import simple_linear_interpolation
```

```python
numerical_imputer = SimpleImputer(strategy = "median")
```

```python
X_train[numericals] = numerical_imputer.fit_transform(X_train[numericals])
```

```python
X_train[numericals].isna().sum()
```

```python
X_test[numericals] = numerical_imputer.transform(X_test[numericals])
```

```python
X_train
```

```python
encoder = OrdinalEncoder(handle_unknown="use_encoded_value",
                         unknown_value=-1)
categorical_imputer = SimpleImputer(strategy= "most_frequent")
```

```python
X_train[categories] = categorical_imputer.fit_transform(X_train[categories])
X_test[categories] = categorical_imputer.transform(X_test[categories])
```

```python
X_train[categories] = encoder.fit_transform(X_train[categories])
X_test[categories] = encoder.transform(X_test[categories])
```

# Modelin Kurulması

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, predictions, squared = False)
```
